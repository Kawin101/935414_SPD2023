1.1	การกระจายตัวแบบเอกรูป (uniform distribution)

```python
# การกระจายตัวแบบเอกรูป (uniform distribution)
#สร้างเลขสุ่มและเขียนภาพการกระจายตัวของค่าสุ่ม X ในช่วง 0 ถึง 1
import numpy as np
import matplotlib.pyplot as plt

X = np.random.uniform(0.0,1.0,10000)
Y = np.random.uniform(0.0,1.0,10000)

plt.plot(X,Y, '*')
plt.show()
```

1.2	การกระจายตัวแบบปกติ (normal distribution)

```python
#การกระจายตัวแบบปกติ (normal distribution)
#สร้างเลขสุ่มและเขียนภาพการกระจายของค่าสุ่ม X ที่มีค่าเฉลี่ย 0.0 และส่วนเบี่ยงเบนมาตรฐาน 0.5
import numpy as np
import matplotlib.pyplot as plt

X0 = np.random.normal(0.0,0.5,10000)
Y0 = np.random.normal(0.0,0.5,10000)

X1 = np.random.normal(10.0,0.5,10000)
Y1 = np.random.normal(10.0,0.5,10000)

X2 = np.random.normal(5.0,1.0,10000)
Y2 = np.random.normal(5.0,1.0,10000)

plt.plot(X0,Y0, 'r+')
plt.plot(X1,Y1, 'bo')
plt.plot(X2,Y2, 'g*')
plt.show()
```

1.3	การกระจายตัวแบบทวินาม (binomial distribution)
<p> โยนลูกเต๋า 1 ลูกจำนวน 100 ครั้งหาความน่าจะเป็นที่จะเกิดแต้มที่มากกว่า 4 เป็นจำนวนครั้งต่างๆ

```python
#การกระจายตัวแบบทวินาม (binomial distribution)
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
#ลูกเต๋ามีแต้มมากกว่า 4 คือ 5 กับ 6 รวมจำนวน 2 แบบ จาก 6 แบบ คิดเป็นความน่าจะเป็นได้ 2/6
sample = np.random.binomial(n = 365, p =0.01, size = 10)
sample = np.round(sample,0)
print(sample)
tt = np.array(range(min(sample),max(sample)+1))
print(tt)
plt.hist(sample, bins=tt, edgecolor = "black")
plt.show()
#ค่าเฉลี่ยการเกิดเหตุการณ์
ave = np.average(sample)
cost = ave*1500
print('จำนวนความบกพร่องที่พบเฉลี่ย', ave, 'เงินสำรองสำหรับซ่อมบำรุง', cost)
```

1.4	การกระจายตัวแบบเอ็กซ์โปเนนเซียล (exponential)
<p> ที่เครื่อง ATM แห่งหนึ่งมีผู้มาใช้งานในอัตรา 20 คน ต่อหนึ่งชั่วโมง หากลูกค้าคนหนึ่งพึ่งมาถึงความน่าจะเป็นที่ลูกค้าคนต่อไปจะมาถึง ณ นาทีต่างๆ ถัดไป ตั้งแต่ 1 ถึง 60 เป็นเท่าไร

```python
#การกระจายตัวแบบเอ็กซ์โปเนนเซียล (exponential)
import numpy as np
import matplotlib.pyplot as plt
#คำนวนค่า beta = 1/lamda, โดยที่ lamda = 20/60 (หนึ่งชั่วดมง = 60 นาที)
time_interval = np.random.exponential(scale=3.0, size=10000)
#print(time_interval)
tt = np.linspace(min(time_interval),max(time_interval),50)
plt.hist(time_interval, bins=tt, edgecolor = "black")
plt.show()
#โอกาสที่ลูกค้าคนถัดไปจะมาถึงในช่วงเวลา 4.0-5.0 นาที
sum4to5 = 0
for t in time_interval:
  if t >= 4.0 and t<5.0:
    sum4to5 = sum4to5+1
Pt = sum4to5/len(time_interval)
print(Pt)
```

3.1 กำหนดตารางข้อมูล x-y จงหาสมการเส้นแนวโน้มข้อมูล y = ax + b โดยใช้การการสุ่ม
<p> x =	[1,	3,	5,	7,	10,	12,	13,	16,	18,	20]
<p>y	= [4,	5,	6,	5,	8,	7,	6,	9,	12,	11]


```python
# Optimization (การหาค่าเหมาะสม)
import numpy as np
import matplotlib.pyplot as plt

xi = [1, 3, 5, 7, 10, 12, 13, 16, 18, 20]
yi = [4, 5, 6, 5, 8, 7, 6, 9, 12, 11]
xi = np.array(xi)
yi = np.array(yi)

# ต้องการหาค่า a, b ที่ทำให้ได้ สมการเส้นตรง ym = axi+b ที่ดีที่สุด (ym ใกล้ yi ที่สุด)
# สุ่มหาค่า a,b
err = np.sum(yi)
a_best = 0.4
b_best = 3.0
print('it == -1', a_best, b_best, err)
for it in range(1000000):
 a = np.random.uniform(0.3, 0.6)
 b = np.random.uniform(2.0, 5.0)
 ym = a*xi + b
 err_it = np.sum((yi-ym)**2)
 if err_it < err:
    a_best = a
    b_best = b
    err = err_it
    print(it, a_best, b_best, err)
plt.plot(xi,yi,'r*')
plt.plot(xi,ym,'bo')
plt.show()
```

3.2 หาค่าเหมาะสมที่สุดสำหรับตัวแบบ SIR โดยใช้ข้อมูล covid_dat.csv

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('covid_dat.csv')
#print(df.head())
#print(df.tail())
#dataI = df('I')
#dataR = df('R')
#print(dataI)
# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.

# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 1.0, 1.0/7.0
beta_best = beta
gamma_best = gamma
errB = 1000000
for ir in range(1000):
# A grid of time points (in days)
  t = np.linspace(0, 70, 70)
  ret = odeint(deriv, y0, t, args=(N, beta, gamma))
  S, I, R = ret.T
  errI = df['I']-I
  errR = df['R']-R
  errM = np.sum(errI**2+errR**2)/len(df['I'])**2
  if errM < errB:
    errB = errM
    beta_best = beta
    gamma_best = gamma
    print(beta_best, gamma_best, 1.0/gamma_best)
  beta  = np.random.uniform(0.0,2.0)
  gamma = 1.0/np.random.uniform(1,15)

#After optimizion
t = np.linspace(0, 150, 150)
ret = odeint(deriv, y0, t, args=(N, beta_best, gamma_best))
S, I, R = ret.T
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
#ax.plot(t, S/1000, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/100, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t[0:70], df['I']/100, 'r+', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/100, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.plot(t[0:70], df['R']/100, 'g*', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (100s)')
ax.set_ylim(0,10)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
