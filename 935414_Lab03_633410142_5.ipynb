1.การหาผลเฉลย (รากของสมการ) ด้วยวิธีแบ่งครึ่งช่วง (bisection method)
<p>1.1 หารากของสมการ $f(x) = 5x^3 - 4x + 1$ ในช่วง [-2, -0.5] โดยมีค่าคลาดเคลื่อนน้อยกว่า 0.1
<p> 1.2 หาค่า $x$ ที่ทำให้ $f(x) = 5 = 5x^3 - 4x + 1$
จะได้ $0 = 5x^3 - 4x - 4$


```python
# 1.1
import numpy as np
import matplotlib.pyplot as plt
def fx(x):
  return 5*x**3 - 4*x + 1
a = -2.0
b = -0.5
eps = 0.00001
for itr in range(100):
  fa = fx(a)       #5*a**3 - 4*a + 1
  fb = fx(b)       #5*b**3 - 4*b + 1
  m = (a + b)/2.0
  fm = fx(m)       #5*m**3 - 4*m + 1
  # print(itr,">>", a, m, b)
  print([itr],[a,m,b])
  # print(fa, fm, fb)
  print([fa,fm,fb],"\n")

  if abs(fm) < eps:  # epsilon
    break
  if fa*fm < 0.0 :
    b = m
  else:
    a = m
```

2.การหาผลเฉลย (รากของสมการ) ด้วยวิธีจุดตรึง
<p>จุดตรึง (Fixed Point) คือจุดที่  $x = g(x)$
<p> 2.1 หารากของสมการ $f(x) = 2x^2 - x - 3 $
<p> $0 = 2x^2 - x - 3 $
<p> $x = 2x^2 - 3 $
<p> $g1(x) = 2x^2 - 3 $
<p> -----------------------
<p>  $x^2 = (x+3)/2$
<p>  $x = sqrt((x+3)/2)$
<p>  $g2(x) = sqrt((x+3)/2)$
<p> 2.2 หาจุดตัดระหว่างเส้นโค้ง $y = cos(2x+1)$ และ เส้นตรง $y = 2x+3$



```python
import numpy as np
import matplotlib.pyplot as plt
import math
# 2.1 : x = 2*x**2-3
x_old = 2.0
print(x_old)
for itr in range (10):
  x_new = 2.0*x_old**2-3.0 #'Numerical result out of range'
  x_old = x_new
  print(x_old)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
#2.1.1: x = sqrt((x+3)/2)
x_old = 2.0
print(x_old)
for itr in range (10):
  x_new = math.sqrt((x_old+3.0)/2.0)
  x_old = x_new
  print(x_old)
```

```python
#2.2 หาจุดตัดระหว่างเส้นโค้ง  y=cos(2x+1)  และ เส้นตรง  y=2x+3
# cos(2x+1) = 2x + 3
# x = (cos(2x+1) - 3)/2 : --> g1(x)
# x = (acos(2x + 3) - 1)/2 : --> g2(x)
import numpy as np
import matplotlib.pyplot as plt
import math
x_old = -3.0
print("x_old =",x_old)
for itr in range (20):
  x_new = (math.cos(2*x_old+1.0)-3.0)/2.0
  error = abs(x_new-x_old)
  x_old = x_new
  print("iteration =", [itr], "x =", [x_old])
  if error < 0.00001:
    break
```

```python
# https://en.wikipedia.org/wiki/Simpson%27s_rule

# https://youtu.be/LPstDKgUuag?si=qvzFlNYtv0p-h-3n

# 3.2 Simpson's Rules การบ้าน
import numpy as np
def f(x):
    return x**2
a = 1.0
b = 4.0
n = 15
def simpsons_rule(f, a, b, n):
    h = (b - a) / n
    result = f(a) + f(b)

    for i in range(1, n, 2):
        result += 4 * f(a + i * h)

    for i in range(2, n-1, 2):
        result += 2 * f(a + i * h)

    result *= h / 3
    return result
integral = simpsons_rule(f, a, b, n)
print("ค่าการบวกโดยประมาณ:", integral)
```
